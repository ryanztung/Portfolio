```python
# Import libraries

import pandas as pd
```

## Cleaning MVP statistics

```python
# Read MVP data into a dataframe

mvp_data = pd.read_csv('mvps.csv')
mvp_data.head()
```

```python
# Select features relevant to MVP voting

features = ['Player', 'Year', 'Pts Won', 'Pts Max', 'Share']
mvp_data = mvp_data[features]
mvp_data.head(10)
```

## Cleaning general player statistics

```python
# Read general player data into a dataframe

player_data = pd.read_csv('players.csv')
player_data.head()
```

```python
player_data.columns
```

```python
# Remove extraneous columns

player_data = player_data.drop(columns=['Unnamed: 0', 'Rk'])
player_data.head()
```

```python
# Remove extraneous characters attached to player names

player_data['Player'] = player_data['Player'].str.replace('*', '', regex=False)
player_data.head(25)
```

### Combine multi-team players

```python
# Locate players associated with multiple teams in one year

df = player_data.groupby(['Player', 'Year'])
df.get_group(('Greg Anderson', 1991))
```

```python
# Reduce multi-team players to an aggregate row for a single year

def TotalStats(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,3]
        return row
    
df = df.apply(TotalStats)
```

```python
df.head(20)
```

```python
# Drop extra indices

df.index = df.index.droplevel()
df.index = df.index.droplevel()
df
```

```python
df[df['Player'] == 'Greg Anderson']
```

## Merge MVP and general player statistics

```python
df = df.merge(mvp_data, on=['Player', 'Year'], how='outer')
```

```python
df.head()
```

```python
df[df['Player'] == 'Stephen Curry']
```

```python
# Fill null values of players without MVP votes

df[['Pts Won', 'Pts Max', 'Share']] = df[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```

```python
df.head(10)
```

## Cleaning team statistics

```python
# Read team data into a dataframe

team_data = pd.read_csv('teams.csv')
team_data.head(15)
```

```python
# Remove categorization rows

team_data = team_data[~team_data['W'].str.contains('Division')]
team_data.head(10)
```

```python
# Clean extraneous characters

team_data['Team'] = team_data['Team'].str.replace('*', '', regex=False)
team_data.head()
```

```python
team_data['GB'] = team_data['GB'].str.replace('—', '0')
team_data.head()
```

## Merge player and team statistics

```python
# Create dictionary that maps abbreviated team name to full team name

abbreviations = {}

with open('nicknames.csv', 'r') as file:
    file.readline().strip()
    
    for line in file:
        abbreviation, name = line.split(',')
        name = name.rstrip('\n')
        abbreviations[abbreviation] = name
```

```python
abbreviations
```

```python
# Map abbreviations dictionary

df['Team'] = df['Tm'].map(abbreviations)
df.head()
```

```python
# Merge team and player datasets

df = df.merge(team_data, on=['Team', 'Year'], how='outer')
df
```

```python
# Drop extraneous columns

df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
# Convert numeric data do numeric datatypes

df.dtypes
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
df.dtypes
```

```python
# Export cleaned dataset to CSV

df.to_csv('cleaned_data.csv')
```
